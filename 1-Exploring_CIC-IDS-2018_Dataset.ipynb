# Exploring the CIC-IDS-2018 Dataset
[official dataset link](https://www.unb.ca/cic/datasets/ids-2018.html)


Based on the following exploration; The __"Friday-02-03-2018_TrafficForML_CICFlowMeter.csv"__ and __"Friday-16-02-2018_TrafficForML_CICFlowMeter.csv"__ (friday 1 and friday 2) csv files were chosen with the following labels:
* __Benign__ (1,209,156 total instances)
* __Bot__ (286,191 total instances)
* __DoS attacks-SlowHTTPTest__ (139,890 total instances)
* __DoS attacks-Hulk__ (461,912 total instances)
* Label (one erroneous row, would need to be removed)

__Reasoning__: The labels are better balanced than the rest of the files. For the purposes of a baseline experiment; it would be satisfactory.

The __Bot__, __DoS attacks-SlowHTTPTest__, and __DoS attacks-Hulk__ labels would be combined into one __"Malicious"__ label for a total of 887,993 __"Malicious"__ instances

The baseline experiment would be a binary classification experiment between __887,993 "Malicious"__ instances vs __1,209,156 "Benign"__ instances.

### 1- Friday logs

```python
import pandas as pd
```

#### 1.1- Friday-02-03-2018_TrafficForML_CICFlowMeter.csv

```python
df_friday1 = pd.read_csv("CIC-IDS-2018-Dataset/Friday-02-03-2018_TrafficForML_CICFlowMeter.csv")
```

```python
df_friday1
```

```python
df_friday1["Label"].unique()
```

```python
print(df_friday1["Label"].value_counts()[['Benign']].sum())
print(df_friday1["Label"].value_counts()[['Bot']].sum())
```
